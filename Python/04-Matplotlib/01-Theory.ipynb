```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#First Lecture
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Function
plt.plot(x,y)
plt.xlabel('Abscissa')
plt.ylabel('Ordinate')
plt.title('Function')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
#OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('Abscissa')
axes.set_ylabel('Ordinate')
axes.set_title('Function')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger')

axes2.plot(y,x)
axes2.set_title('Smaller')
```

```python
#Second Lecture
```

```python
# რა განსხვავებაა figure() და subplots() შორის , subplots() თავისით ქმნის axes დიაპაზონებს
# axes რეალურად არის array რომელზე იტერაციაც შეგვიძლია
fig,axes = plt.subplots(nrows = 1 , ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1 , ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')


axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
#Figure size and DPI
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize=(3,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure(figsize = (3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label ='X Cubed')

ax.legend(loc=0)
```

```python
#Third Lecture
```

```python
#linewidth = lw
#linestyle = ly
fig = plt.figure(figsize = (3,2))
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green',lw = 3,alpha =0.6,ls = '--')
```

```python
#markersize = ms
fig = plt.figure(figsize = (4,3))
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'green',lw = 2.5,ls = '-',marker = 'o',ms = 6)
```

```python
fig = plt.figure(figsize = (4,3))
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',lw = 2.5,ls = '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
